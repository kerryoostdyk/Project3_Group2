```python
import pandas as pd
import numpy as no
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine, inspect, text, func
```

```python
engine = create_engine("sqlite:///UFO_data.sqlite")
```

```python
inspector = inspect(engine)
```

```python
tables = inspector.get_table_names()

for table in tables:
	print(table)
	print("--------")
	columns = inspector.get_columns(table)
	for column in columns: 
		print(column['name'], column['type'])

	print()
```

```python
conn = engine.connect()
```

```python
query = text("SELECT * FROM UFO_data;")

df1 = pd.read_sql(query, con=conn)
df1.head()
```

```python
query = text("SELECT UFO_shape, COUNT(UFO_shape) FROM UFO_data GROUP BY UFO_shape;")

shape_df = pd.read_sql(query, con=conn)
shape_df = shape_df.rename(columns={'COUNT(UFO_shape)' : 'Number of Encounters'})
shape_df.head()
```

```python
shape_df = shape_df.drop(index=0) 
shape_df.head()
```
